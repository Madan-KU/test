```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

data =df
# # Load the data
# data = pd.read_csv(r"C:\Users\Home\Desktop\Madan 2023\X\00.Interview\JLR\WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
data.transpose()
```

```python
data["Attrition"] = LabelEncoder().fit_transform(data['Attrition'])
data["BusinessTravel"] = LabelEncoder().fit_transform(data['BusinessTravel'])
data["Department"] = LabelEncoder().fit_transform(data['Department'])
data["EducationField"] = LabelEncoder().fit_transform(data['EducationField'])
data["Gender"] = LabelEncoder().fit_transform(data['Gender'])
data["JobRole"] = LabelEncoder().fit_transform(data['JobRole'])
data["MaritalStatus"] = LabelEncoder().fit_transform(data['MaritalStatus'])
data["Over18"] = LabelEncoder().fit_transform(data['Over18'])
data["OverTime"] = LabelEncoder().fit_transform(data['OverTime'])
data["Education"] = LabelEncoder().fit_transform(data['Education'])
data["JobInvolvement"] = LabelEncoder().fit_transform(data['JobInvolvement'])
data["JobSatisfaction"] = LabelEncoder().fit_transform(data['JobSatisfaction'])
data["PerformanceRating"] = LabelEncoder().fit_transform(data['PerformanceRating'])
data["RelationshipSatisfaction"] = LabelEncoder().fit_transform(data['RelationshipSatisfaction'])
data["WorkLifeBalance"] = LabelEncoder().fit_transform(data['WorkLifeBalance'])
data["EnvironmentSatisfaction"] = LabelEncoder().fit_transform(data['EnvironmentSatisfaction'])
```

```python
X = data.drop(['Attrition'], axis=1)
y = data['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
# Create a logistic regression model 
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
```

```python
coefficients = pd.DataFrame(model.coef_, columns=X.columns)

equation = 'y = '
for col in X.columns:
    equation += '{:.5f} * {} + '.format(coefficients[col].values[0], col)
equation += '{:.5f}'.format(model.intercept_[0])
print(equation)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy score:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier

# Create the random forest classifier with the best parameters
random_forest = RandomForestClassifier(criterion='gini', max_depth=7, max_features='auto', n_estimators=100)

# Fit the model to your data
random_forest.fit(X_train, y_train)

# Predict on the training and test sets
y_train_pred = random_forest.predict(X_train)
y_test_pred = random_forest.predict(X_test)

# Calculate the accuracy scores
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print the accuracy scores
print("Train Accuracy : {:.2f} %".format(train_accuracy*100))
print("Test Accuracy : {:.2f} %".format(test_accuracy*100))

# Calculate the confusion matrix for the test set
confusion_mat = confusion_matrix(y_test, y_test_pred)

# Plot the confusion matrix
import seaborn as sns
sns.heatmap(confusion_mat, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=7, max_features='sqrt')


num_features = 10


rfe = RFE(estimator=random_forest, n_features_to_select=num_features)


rfe.fit(X_train, y_train)


selected_features = X_train.columns[rfe.support_]
print("Selected Features: ", selected_features)


random_forest.fit(X_train[selected_features], y_train)
train_accuracy = random_forest.score(X_train[selected_features], y_train)
test_accuracy = random_forest.score(X_test[selected_features], y_test)
print("Train Accuracy : {:.2f} %".format(train_accuracy*100))
print("Test Accuracy : {:.2f} %".format(test_accuracy*100))

# Confusion Matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = random_forest.predict(X_test[selected_features])
cm = confusion_matrix(y_test, y_pred)
classes = ["0", "1"]
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', xticklabels=classes, yticklabels=classes)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
import matplotlib.pyplot as plt

importances = random_forest.feature_importances_
feature_names = list(X_train.columns)

feature_importances = dict(zip(feature_names, importances))

sorted_feature_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)

top_n = 10  #n Features

plt.bar(range(top_n), [importance for _, importance in sorted_feature_importances[:top_n]])
plt.xticks(range(top_n), [feature for feature, _ in sorted_feature_importances[:top_n]], rotation=90)
plt.xlabel("Feature")
plt.ylabel("Importance Score")
plt.title("Top {} Feature Importance Scores".format(top_n))
plt.show()
```
